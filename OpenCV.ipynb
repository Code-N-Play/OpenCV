```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Images /img3.jpeg")
resize_img = cv2.resize(img, (250,300))


# hight line

line = cv2.line(img = resize_img, # image jispr line put krni hai
                pt1 = (220,60),   # first point of line (x,y)
                pt2 = (220,250),  # Second point of line (x,y)
                color = (0,0,255),# line color
                thickness = 2,    # line thikness
                lineType = cv2.LINE_4,)

# hight text

text = cv2.putText(line,
                   text = "5.1",
                   org = (222,160),
                   fontFace = cv2.FONT_HERSHEY_COMPLEX_SMALL,
                   fontScale= 1,
                   color = (255,0,0),
                   thickness = 1,
                   lineType = 1 ,
                   bottomLeftOrigin = False)

cv2_imshow(text)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Images /img4.jpg")
resize_img = cv2.resize(img,(600,400))

box = cv2.rectangle(img = resize_img,
                    pt1 = (200,60),
                    pt2 = (500,290),
                    color = (0,255,0),
                    thickness = 2,
                    lineType = 4 )

text = cv2.putText(img = box,
                   text = "Oggy",
                   org = (200,50),
                   fontFace = 1,
                   fontScale = 2.1,
                   color = (0,0,255),
                   thickness = 2,
                   lineType = 4,
                   bottomLeftOrigin  = False)

cv2_imshow(text)
```
